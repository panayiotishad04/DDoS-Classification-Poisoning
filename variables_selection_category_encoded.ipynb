```python
df_ben = df_ben.drop(['ts_time','uid_string','service_string','orig_bytes_count','resp_bytes_count','local_orig_bool','local_resp_bool','missed_bytes_count','tunnel','Category-detail','history_string'], axis=1)
df_ddos = df_ddos.drop(['ts_time','uid_string','service_string','orig_bytes_count','resp_bytes_count','local_orig_bool','local_resp_bool','missed_bytes_count','tunnel','Category-detail','history_string'], axis=1)
```

```python
print("Shape benigns",df_ben.shape)
print("shape ddos",df_ddos.shape)
```

```python
# Concatenate ddos and ben in one single database to transform the categorical fields
df_ben_ddos = pd.concat([df_ben,df_ddos],ignore_index=True)
df_ben_ddos
```

```python
# Clean the 'duration_interval' column
df_ben_ddos['duration_interval'] = df_ben_ddos['duration_interval'].str.replace('-', ' ')
df_ben_ddos['duration_interval'] = pd.to_numeric(df_ben_ddos['duration_interval'], errors='coerce')
df_ben_ddos[['duration_interval']] = df_ben_ddos[['duration_interval']].fillna(value=df_ben_ddos[['duration_interval']].mean())
```

```python
# Categorical fields: id.orig_addr, id.resp_haddr, proto_enum, conn_state_string, Category
import ipaddress

# Convert IP addresses to string type
# df_ben_ddos['id.orig_addr'] = df_ben_ddos['id.orig_addr'].astype('string')
# df_ben_ddos['id.resp_haddr'] = df_ben_ddos['id.resp_haddr'].astype('string')


# Loop to transform categorical fields to numbers
categorical_col = ['proto_enum', 'conn_state_string', 'Category']
mapping = {}

for col in categorical_col:
    le = LabelEncoder()     # Initialize LabelEncoder
    df_ben_ddos[col] = le.fit_transform(df_ben_ddos[col])
    mapping[col] = le

# Transform IP addresses to numbers
ip_col = ['id.orig_addr','id.resp_haddr']
for col in ip_col:
    df_ben_ddos[ip_col] = df_ben_ddos[ip_col].map(lambda ip: int(ipaddress.IPv4Address(ip)))


map_used = ['proto_enum', 'conn_state_string', 'Category']
for col in map_used:
    print(f"\nMapeo utilizado para la columna '{col}':")
    for idx, category in enumerate(mapping[col].classes_):
        print(f"{category} -> {idx}")

df_ben_ddos
```
